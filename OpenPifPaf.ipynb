```python
def get_angle(edge1,  edge2):
    edge1 = set(edge1)
    edge2 = set(edge2)
    mid_point = edge1.intersection(edge2).pop()
    a = (edge1-edge2).pop()
    b = (edge2-edge1).pop()
    v1 = point[mid_point]-point[a]
    v2 = point[mid_point]-point[b]

    angle = (math.degrees(np.arccos(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))))
    return angle
```
